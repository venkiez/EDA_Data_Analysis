# WEB SCRAPPING Project on Analysing Used Car

- Web scraping is an automated method to extract lage amounts of data from websites.

**Basic Steps:**

1. Identify the URL from which you need the data
2. Inspect the code behind the page
3. Find the elements/data you want to extract
4. Write and excute the code to store the data in the required format

**Python libraries required:**

* **`requests:`** This is used to extract the html code from the give URL.
* **`BeautifulSoup:`** This is used to format and read the html content.
* **`re:`** To handle the text data.
* **`numpy & pandas`**

## Identify the outliers

```python
# outerliers can be identified using boxplot method and z-score method.
# Checking the outliers using the boxplot method.
# 1. Price Column:

Q1,Q2,Q3 = tuple(Car_df.Total_price.quantile(q=[0.25,0.50,0.75]).values)

# Checking the missing values

print(Q1,Q2,Q3)
```

```python
# Calculating the IQR value and for calculating the LL and UL.
IQR = Q3-Q1

# Checking the outliers in lowerlimit(LL).`
UL = Q3+1.5*IQR

# Checking the outliers in upperlimit(UL)
LL = Q1-1.5*IQR

print(IQR,UL,LL)
```

```python
Car_df[(Car_df.Total_price < LL) | (Car_df.Total_price > UL)]
```
